Step 4: Start Data Cleaning

```python
if df is not None:
    print("--- Handling Date Columns ---")
    date_columns = ['Date_received', 'Date_sent']
    existing_date_cols = []

    for col in date_columns:
        if col in df.columns:
            print(f"\nProcessing column: {col}")
            original_nulls = df[col].isnull().sum()
            df[col] = pd.to_datetime(df[col], dayfirst=True, errors='coerce')

            new_nulls = df[col].isnull().sum()
            if new_nulls > original_nulls:
                print(f"Warning: Some values in '{col}' could not be parsed as dates and are now NaT (missing). Additional NaNs: {new_nulls - original_nulls}")

            print(f"Sample of parsed dates in '{col}':")
            print(df[col].dropna().head())
            print(f"Column '{col}' converted to datetime objects.")
            existing_date_cols.append(col)
        else:
            print(f"Warning: Date column '{col}' not found. Skipping date conversion for this column.")

    if existing_date_cols:
        print("\n--- Data types of date columns after conversion ---")
        print(df[existing_date_cols].info())
        print("\n--- Quick check of min/max for parsed date columns ---")
        for col in existing_date_cols:
            if col in df.columns and df[col].notna().any():
                 print(f"{col}: Min Date = {df[col].min()}, Max Date = {df[col].max()}")

    else:
        print("\nNo specified date columns were found or processed.")

    print("\nNote: Dates are now stored as datetime objects. ")
    print("For specific string formatting (e.g., 'dd-mm-YYYY') for final output or display, ")
    print("you would convert them back to strings using .dt.strftime('%d-%m-%Y') at a later stage if needed.")

else:
    print("DataFrame not loaded. Skipping date column handling.")
```

```python
if df is not None:
    print("\n--- Handling Missing Values in Categorical Columns ---")
    categorical_fill_unknown = ['Sub-product','Sub-issue', 'State']
    for col in categorical_fill_unknown:
        if col in df.columns:
            missing_before = df[col].isnull().sum()
            if missing_before > 0:
                df[col] = df[col].fillna('Unknown')
                print(f"Filled {missing_before} NaNs in '{col}' with 'Unknown'.")
            else:
                print(f"No NaNs to fill in '{col}'.")
        else:
            print(f"Warning: Categorical column '{col}' for NaN filling not found. Skipping.")
else:
    print("DataFrame not loaded. Skipping NaN handling for categorical columns.")
```

```python
if df is not None:
    print("\n--- Handling Missing Values and Type for ZIP Code Column ---")
    zip_col_name = 'ZIP'
    if zip_col_name in df.columns:
        missing_before = df[zip_col_name].isnull().sum()
        df[zip_col_name] = df[zip_col_name].astype(str).fillna('00000')
        print(f"Ensured '{zip_col_name}' is string type. Filled {missing_before} (original) NaNs with '00000'.")
    else:
        zip_col_name_alt = 'ZIP code'
        if zip_col_name_alt in df.columns:
            zip_col_name = zip_col_name_alt
            missing_before = df[zip_col_name].isnull().sum()
            df[zip_col_name] = df[zip_col_name].astype(str).fillna('00000')
            print(f"Found column as '{zip_col_name}'. Ensured it is string type. Filled {missing_before} (original) NaNs with '00000'.")
        else:
            print(f"Warning: ZIP code column ('{zip_col_name}' or '{zip_col_name_alt}') not found. Skipping ZIP code cleaning.")
            zip_col_name = None
else:
    print("DataFrame not loaded. Skipping ZIP code cleaning.")
    zip_col_name = None
```

```python
if df is not None:
    print("\n--- Cleaning 'Disputed' Column ---")
    disputed_col_name = 'Consumer disputed?'
    if disputed_col_name in df.columns:
        print(f"Unique values in '{disputed_col_name}' before cleaning: {df[disputed_col_name].unique()}")
        missing_before = df[disputed_col_name].isnull().sum()

        df[disputed_col_name] = df[disputed_col_name].fillna('Data Missing')

        df[disputed_col_name] = df[disputed_col_name].astype(str).str.lower().replace('n/a', 'not applicable')

        valid_disputed_values = ['yes', 'no', 'not applicable', 'data missing']

        df[disputed_col_name] = df[disputed_col_name].replace('nan', 'data missing')

        df[disputed_col_name] = df[disputed_col_name].apply(lambda x: x if x in valid_disputed_values else 'data missing')

        print(f"Filled/standardized NaNs in '{disputed_col_name}' (originally {missing_before} NaNs).")
        print(f"Unique values in '{disputed_col_name}' after cleaning: {df[disputed_col_name].unique()}")
    else:
        print(f"Warning: Column '{disputed_col_name}' not found. Skipping 'Disputed' cleaning.")
else:
    print("DataFrame not loaded. Skipping 'Disputed' column cleaning.")
```

```python
if df is not None:
    print("\n--- Checking for and Removing Duplicate Rows ---")
    initial_row_count = len(df)
    df.drop_duplicates(inplace=True)
    rows_dropped = initial_row_count - len(df)
    if rows_dropped > 0:
        print(f"Removed {rows_dropped} duplicate rows.")
    else:
        print("No duplicate rows found.")
    print(f"Current number of rows: {len(df)}")
else:
    print("DataFrame not loaded. Skipping duplicate row check.")
```
